# Quiz 9 วนลูปนับว่านิทานที่มีคำว่า แม่หรือพ่อ กี่เรื่อง เรื่องอะไรบ้าง

# มีคำดังกล่าวในนิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน

# คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด

````
# prompt:  วนลูปนับว่านิทานที่มีคำว่า แม่หรือพ่อ กี่เรื่อง เรื่องอะไรบ้าง
# มีคำดังกล่าวในนิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
#  คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด

count_mom_dad = 0 #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อ
list_mom_dad_story = [] #เก็บชื่อนิทานที่มีคำว่าแม่หรือพ่อ
count_mom_dad_life = 0 #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อในนิทานชีวิต
count_mom_dad_local = 0 #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อในนิทานประจำถิ่น
dict_creator = {} #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อ
for i, each_row in this_data.iterrows(): #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อ
    if 'แม่' in each_row['title'] or 'พ่อ' in each_row['title']: #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อ
        count_mom_dad += 1 #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อ
        list_mom_dad_story.append(each_row['title']) #เก็บชื่อนิทานที่มีคำว่าแม่หรือพ่อ
        if each_row['type'] == 'นิทานชีวิต':  #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อในนิทานชีวิต
            count_mom_dad_life += 1 #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อในนิทานชีวิต
        elif each_row['type'] == 'นิทานประจำถิ่น': #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อในนิทานประจำถิ่น
            count_mom_dad_local += 1 #นับจำนวนนิทานที่มีคำว่าแม่หรือพ่อในนิทานประจำถิ่น
        if each_row['creator'] not in dict_creator: #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อ
            dict_creator[each_row['creator']] = 0 #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อ
        dict_creator[each_row['creator']] += 1 #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อ
print('มีนิทานที่มีคำว่า แม่หรือพ่อ อยู่',count_mom_dad,'เรื่อง')
print('รายชื่อนิทานที่มีคำว่า แม่หรือพ่อ',list_mom_dad_story)
if count_mom_dad_life > count_mom_dad_local: #เช็คว่ามีคำดังกล่าวในนิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
    print('มีคำดังกล่าวในนิทานชีวิตมากกว่า')
elif count_mom_dad_life < count_mom_dad_local: #เช็คว่ามีคำดังกล่าวในนิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
    print('มีคำดังกล่าวในนิทานประจำถิ่นมากกว่า')
else:
    print('มีคำดังกล่าวในนิทานชีวิตและนิทานประจำถิ่นเท่ากัน')
max_count = 0 #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด
max_creator = '' #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด
for creator, count in dict_creator.items(): #นับคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อ
    if count > max_count: #เช็คว่าคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด
        max_count = count #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด
        max_creator = creator #เก็บคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด
print('คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุดคือ',max_creator)
````

````
# prompt: # prompt: บอกเหตุผล 3 ผู้กำกับหนังอินเดียชั้นนำ จาก Rating และ Votes ที่มากกว่า 80% และหนังยอดนิยมของผู้กำกับหนังอินเดียชั้นนำ 3 คน และประเภทของหนัง

# คำนวณค่าเฉลี่ยของ Rating และ Votes
average_rating = movie_df['Rating'].mean()
average_votes = movie_df['Votes'].mean()

# กำหนดเกณฑ์สำหรับผู้กำกับชั้นนำ (Rating และ Votes มากกว่า 80% ของค่าเฉลี่ย)
top_director_threshold_rating = average_rating * 0.8
top_director_threshold_votes = average_votes * 0.8

# เลือกผู้กำกับที่ตรงตามเกณฑ์
top_directors = movie_df.groupby('Director').filter(lambda x: (x['Rating'].mean() > top_director_threshold_rating) and (x['Votes'].mean() > top_director_threshold_votes))

# นับจำนวนหนังของผู้กำกับแต่ละคน
director_counts = top_directors['Director'].value_counts()

# เลือกผู้กำกับ 3 อันดับแรก
top_3_directors = director_counts.head(3)

# พิมพ์ข้อมูลของผู้กำกับชั้นนำ 3 คน
for director, count in top_3_directors.items():
    director_df = top_directors[top_directors['Director'] == director]
    top_rated_movie = director_df.loc[director_df['Rating'].idxmax()]
    print(f"ผู้กำกับ: {director} (จำนวนหนัง: {count})")
    print(f"  - เหตุผล: มีคะแนนเฉลี่ย Rating และ Votes สูงกว่า 80% ของค่าเฉลี่ยทั้งหมด")
    print(f"  - หนังยอดนิยม: {top_rated_movie['Name']} (ประเภท: {top_rated_movie['Genre']}, Rating: {top_rated_movie['Rating']})")
    print()
````

1. ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/Dataviz_2024_DATA/test-TidyData_2024.xlsx'
TidyData = pd.read_excel(file_path)
TidyData.head()
```

```python
# clean data
# ลบค่า null ที่มีในทุก columns
TidyData_cleaned = TidyData.dropna(how='all')
TidyData_cleaned
```

```python
#ดึงปีออกจากคอลัมน์ month แล้วเพิ่มเป็นคอลัมน์ใหม่
TidyData_cleaned['year'] = TidyData['month'].str.extract(r'(\d{4})').astype(int)
TidyData_cleaned
```

```python
#คำนวณค่าเฉลี่ยของ Chlo_a สำหรับแต่ละ station ในปี 2018 และ 2019
mean_chlo_a_2018 = TidyData_cleaned[TidyData_cleaned['year'] == 2018].groupby('station')['Chlo_a'].mean()
mean_chlo_a_2019 = TidyData_cleaned[TidyData_cleaned['year'] == 2019].groupby('station')['Chlo_a'].mean()
```

```python
# รวมค่าเฉลี่ยเข้าด้วยกันใน DataFrame เดียวเพื่อทำการเปรียบเทียบ
mean_chlo_a_comparison = pd.DataFrame({
    'Mean Chlo_a 2018': mean_chlo_a_2018,
    'Mean Chlo_a 2019': mean_chlo_a_2019
})
# แสดงผลลัพธ์
mean_chlo_a_comparison
# เพิ่ม columns different ขึ้นมาเพื่อดูความแตกต่าง
mean_chlo_a_comparison = pd.DataFrame({
    'Difference': mean_chlo_a_2019 - mean_chlo_a_2018,
    'Mean Chlo_a 2018': mean_chlo_a_2018,
    'Mean Chlo_a 2019': mean_chlo_a_2019
})
mean_chlo_a_comparison
```

2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
#ใช้คำสั่ง group by จัดกลุ่มช้อมูลตาม station และ month และคำนวณหาค่าเฉลี่ยของ Chlo_b
mean_chlo_b_by_month_station = TidyData_cleaned.groupby(['station', 'month'])['Chlo_b'].mean().reset_index()
mean_chlo_b_by_month_station
```

```python
# ใช้ pivot เพื่อปรับรูปแบบตาราง เพื่อให้ station เป็นแถว และ month เป็น columns
mean_chlo_b_table = mean_chlo_b_by_month_station.pivot(index='station', columns='month', values='Chlo_b')
# แสดงในรูปแบบของตาราง
mean_chlo_b_table
```

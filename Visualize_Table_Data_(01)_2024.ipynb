```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates

```
pd.plotting.parallel_coordinates(df, 'Name'); # เอา columns ที่เราต้องการจะให้มันแยกสี
```

## ปรับ parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#a9d963', '#ee9df1', '#f1c79d')) #html color
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r', 'g', 'b')) #ตัวย่อของสี R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.8,0), (0,0.7,0), (0.6,0,1))) #rgb value (r,g,b) | r,g,b -> [0,1]
```

## ปรับ parameter: colormap

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='Paired') # colormap คือสีที่คิดมาแล้วให้มา map อยู่ในกราฟเดียวกัน
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

# เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy() # แปลง data จาก pandas dataframe เป็น numpy array
np_data # แสดงผล
```

# Visualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของ figure ถ้าไม่กำหนดก็จะเล็ก

plt.imshow(np_data[:50,:],interpolation='nearest') # เราเอา data มาโชว์ในรูปแบบรูปภาพ
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='plasma') # cmap คือ colormap
```

# การใช้ subplot

```
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่) การนำกราฟหลายๆกราฟมาแปะใน plot เดียวกัน เป็นสิ่งที่สำคัญที่สุดในการplot
plt.imshow(np_data[:50,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,2)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma')
```

```
plt.imshow(np_data[:10,:],interpolation='nearest',cmap='magma',alpha = 1 ) # alpha คือ chanel ที่บอกว่าสีโปร่งใสแค่ไหน
```
